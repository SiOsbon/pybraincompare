# How much information is lost at different voxel dimensions?
This notebook will test how pearson similarity scores change when we vary the voxel dimension. The "base" or "gold standard" comparison will be done at a 2mm voxel, and we will vary the threshold up to 9mm.  We will use sum of squared distances as our error, and the goal is to be able to report to users of pybraincompare how much "information is lost" by calculating the comparison at a larger voxel size.

```python
import pandas

# We will test with 10 cases, real analysis will use 100
gs = pandas.read_csv("gs_voxel_comparison_10.tsv",sep="\t",index_col=0) # this is 2mm
resampled = pandas.read_csv("resampled_voxel_comparison_10.tsv",sep="\t",index_col=0)

# For each of the resampled, save sum of squared errors
voxdims = resampled.voxdim.unique()
errors = pandas.DataFrame(columns=gs.columns)
for voxdim in voxdims:
    subset = resampled[resampled.voxdim==voxdim]
    subset.index = [x.replace("_%s" %(voxdim[1]),"") for x in subset.index]
    # ensure same arrangement
    subset = subset.loc[gs.index,gs.columns]
    difference = (gs-subset)
    sq_difference = difference*difference
    errors.loc[voxdim] = sq_difference.sum().tolist()
```

The above is the mean sum of squared differences for each voxel dimension as compared to the gold standard (the original images compared at 2mm).
